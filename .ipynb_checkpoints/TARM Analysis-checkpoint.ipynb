```python
import lyricsgenius
import nltk
import os
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
mySecretToken = "use_yours_here"
artistName = 'Tre allegri ragazzi morti'
```

```python
nltk.download('stopwords')
```

```python
if(os.path.isfile('Lyrics_'+artistName.replace(' ','')+'.json')):
    print('File already exists')
else:
    genius = lyricsgenius.Genius(mySecretToken)
    artist = genius.search_artist(artistName, sort = 'title')
    artist.save_lyrics()
```

```python
with open('Lyrics_'+artistName.replace(' ','')+'.json') as f:
    data = json.load(f)
```

```python
words = []
for song in data['songs']:
    lyrics = song['lyrics'].replace('\n',' ').lower()
    words = words + nltk.tokenize.word_tokenize(lyrics) 
len(words)
```

```python
from nltk.corpus import stopwords
words = [x for x in words if x not in stopwords.words('italian')+[',','.','...','(',')','?','’',"'",'"',':','[',']','``','!']]
len(words)
fdist = nltk.probability.FreqDist(words)
```

```python
df_fdist = pd.DataFrame.from_dict(fdist, orient='index', columns=['Frequency'])
pd.set_option('display.max_rows', 3000)
df_fdist.sort_values(by='Frequency', ascending=False)
```
